```python
from tg_config import TG_TOKEN
import telebot
import random
from telebot import types

TOKEN = TG_TOKEN
count = [0, 0]


bot = telebot.TeleBot(TOKEN)



def get_random(num):
    return random.randrange(num)

def count_add(a, b):
    count[0] += a
    count[1] +=b
    
  

@bot.message_handler(commands=["start"])
def start(m, res=False):
    # –î–æ–±–∞–≤–ª—è–µ–º –∫–Ω–æ–ø–∫–∏
    markup=types.ReplyKeyboardMarkup(resize_keyboard=True)
    
 
    # TODO –¥–æ–±–∞–≤–∏—Ç—å 3 –æ–±—ä–µ–∫—Ç–∞ KeyboardButton –¥–ª—è '–ö–∞–º–µ–Ω—å','–ù–æ–∂–Ω–∏—Ü—ã','–ë—É–º–∞–≥–∞'
    markup = types.ReplyKeyboardMarkup(row_width=3)
    itembtn1 = types.KeyboardButton('–ö–∞–º–µ–Ω—å')
    itembtn2 = types.KeyboardButton('–ù–æ–∂–Ω–∏—Ü—ã')
    itembtn3 = types.KeyboardButton('–ë—É–º–∞–≥–∞')

     # TODO –¥–æ–±–∞–≤–∏—Ç—å –∫–Ω–æ–ø–∫–∏ –≤ –∫–ª–∞–≤–∏–∞—Ç—É—Ä—É (markup)
    markup.add(itembtn1, itembtn2, itembtn3)

    bot.send_message(m.chat.id, '–ù–∞–∂–º–∏ –∫–Ω–æ–ø–∫—É –∏ –Ω–∞—á–Ω–∏ –∏–≥—Ä—É ',  reply_markup=markup)
        

@bot.message_handler(content_types=["text"])
def handle_text(message):
    form = ['–ö–∞–º–µ–Ω—å','–ù–æ–∂–Ω–∏—Ü—ã','–ë—É–º–∞–≥–∞']
    answer = form[get_random(3)]
    
    if count[0] >9:
        count[0] -=count[0]
        count[1]-=count[1]
        bot.send_message(message.chat.id,  '*–ü–æ–∑–¥—Ä–∞–≤–ª—è—é! –¢—ã –≤—ã–∏–≥—Ä–∞–ª* üëç', parse_mode="Markdown")
    elif count[1] >9:
        count[0] -=count[0]
        count[1]-=count[1]
        bot.send_message(message.chat.id,  '*–ú–Ω–µ –∂–∞–ª—å, –Ω–æ —Ç—ã –ø—Ä–æ–∏–≥—Ä–∞–ª!* üëé', parse_mode="Markdown")
    
    # TODO –ø–æ—Å—ã–ª–∞—Ç—å –≤ —á–∞—Ç —Å–ª—É—á–∞–π–Ω–æ–µ –∏–∑ —Å–ø–∏—Å–∫–∞ '–ö–∞–º–µ–Ω—å','–ù–æ–∂–Ω–∏—Ü—ã','–ë—É–º–∞–≥–∞'
    
    answer = form[get_random(3)]
    
    #—Å–º–µ–Ω–∞ —Å—á–µ—Ç–∞
    if message.text == answer :
        message_for_user = '–ù–∏—á—å—è'
    elif message.text == '–ö–∞–º–µ–Ω—å' and answer == '–ù–æ–∂–Ω–∏—Ü—ã' :
        count_add(1, 0)
        message_for_user = '–ö–∞–º–µ–Ω—å —Å–æ–∫—Ä—É—à–∞–µ—Ç –Ω–æ–∂–Ω–∏—Ü—ã'
    elif message.text == '–ö–∞–º–µ–Ω—å' and answer == '–ë—É–º–∞–≥–∞' :
        message_for_user = '–ë—É–º–∞–≥–∞ –ø–æ–∫—Ä—ã–≤–∞–µ—Ç –∫–∞–º–µ–Ω—å'
        count_add(0, 1)
    elif message.text == '–ù–æ–∂–Ω–∏—Ü—ã' and answer == '–ë—É–º–∞–≥–∞' :
        message_for_user = '–ù–æ–∂–Ω–∏—Ü—ã —Ä–µ–∂—É—Ç –±—É–º–∞–≥—É'
        count_add(1, 0)
    elif message.text == '–ù–æ–∂–Ω–∏—Ü—ã' and answer == '–ö–∞–º–µ–Ω—å' :
        message_for_user = '–ö–∞–º–µ–Ω—å —Å–æ–∫—Ä—É—à–∞–µ—Ç –Ω–æ–∂–Ω–∏—Ü—ã'
        count_add(0, 1)
    elif message.text == '–ë—É–º–∞–≥–∞' and answer == '–ö–∞–º–µ–Ω—å' :
        message_for_user = '–ë—É–º–∞–≥–∞ –ø–æ–∫—Ä—ã–≤–∞–µ—Ç –∫–∞–º–µ–Ω—å'
        count_add(1, 0)
    elif message.text == '–ë—É–º–∞–≥–∞' and answer == '–ù–æ–∂–Ω–∏—Ü—ã' :
        message_for_user = '–ù–æ–∂–Ω–∏—Ü—ã —Ä–µ–∂—É—Ç –±—É–º–∞–≥—É'
        count_add(0, 1)
    else:
        message_for_user = None


    if message_for_user == None:
        bot.send_message(message.chat.id,  '‚ÄºÔ∏è–í–≤–µ–¥–µ–Ω–æ –Ω–µ –≤–∞–ª–∏–¥–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞‚ÄºÔ∏è')
    else:
        bot.reply_to(message,f'''{answer}
        <i>{message_for_user}</i>''', parse_mode="HTML")
 
        
#         markdown = '''
#         *bold text*
#         [text](URL)
#         '''
#         bot.send_message(message.chat.id, f'–°—á–µ—Ç: *{count[0]} : {count[1]}*', parse_mode="Markdown")
        bot.send_message(message.chat.id, f'üìå<i>–°—á–µ—Ç </i> <b>{count[0]} : {count[1]}</b>', parse_mode="HTML") 


bot.polling(none_stop=True, interval=0) 
```
